```python
# Load packages
import os
import pandas as pd
import numpy as np

# This line is needed to display plots inline in Jupyter Notebook
%matplotlib inline

# Required for basic python plotting functionality
import matplotlib.pyplot as plt

# Required for formatting dates later in the case
import datetime
import matplotlib.dates as mdates

# Required to display image inline
from IPython.display import Image

# Advanced plotting functionality with seaborn
import seaborn as sns
sns.set(style="whitegrid") # can set style depending on how you'd like it to look
import locale
```

```python
locale.setlocale(locale.LC_ALL, 'es_ES')
df = pd.read_csv('full_df.csv', sep='\t')

df['Fecha Matrícula']=df['Fecha Matrícula'].apply(lambda x: datetime.datetime.strptime(str(x), '%A, %d de %B de %Y'))
df['Fecha Tramite Posterior']=df['Fecha Tramite Posterior'].apply(lambda x: datetime.datetime.strptime(str(x), '%A, %d de %B de %Y'))
df['diff_months'] = df['Fecha Tramite Posterior'] - df['Fecha Matrícula']
df['diff_months']=df['diff_months']/np.timedelta64(1,'M')
df.columns
```

```python
df_Item=pd.read_csv('Item_Code.csv', sep=';')

df_merged=pd.merge(df, df_Item, on='ItemCode', how='left')
df_merged
```

```python
plt.figure(figsize=(15, 7))
matriculas_mes={}
period = df['Fecha Matrícula'].dt.to_period("m")
matriculas= df.groupby(period)
matriculas_mes['ID']=matriculas['ID'].count().plot()
plt.title('Registration Date of the vehicle');
```

```python
plt.figure(figsize=(15, 7))
Tramites_mes={}
period = df['Fecha Tramite Posterior'].dt.to_period("m")
tramites= df.groupby(period)
Tramites_mes['ID']=tramites['ID'].count().plot()
plt.title('Processing Date');
```

```python
pivot_merged = df_merged.pivot('Fecha Tramite Posterior','Fecha Matrícula','Tipo')

# Heat map of energy consumption by month and year
fig, ax = plt.subplots(figsize=(15,5))
sns.heatmap(pivot_merged, cmap="coolwarm", ax=ax);
ax.set_title('PEC Electric Power Sector');
```
